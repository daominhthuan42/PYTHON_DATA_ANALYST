## <b> Pyplot API vs Object-Oriented (OO) API

- Pyplot API -> Quickly.
- OO API -> Advanced.

### <b> Pyplot API

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), color = "blue", linestyle = "dashed", label = "sin(x)");
plt.plot(x, np.cos(x), color = "red", linestyle = "dashed", label = "cos(x)");
plt.xlabel("Biến X")
plt.ylabel("F(x)")
plt.title("Sin(x) and Cos(x)")
# plt.xlim(0, 4)
# plt.ylim(-0.7, 0.6)
# plt.axis([0, 4, -0.7, 0.6])
plt.legend()
plt.show()
```

### <b> Object-Oriented (OO) API

```python
#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5)) #Figure size = Width & Height of the plot

#3. Plot the data
ax.plot(x, y);

#4. Customize the plot
ax.set(title="A simple plot",
       xlabel="x-axis",
       ylabel="y-axis");
```

### <b> 2. Scatter

```python
# Pylot API
plt.scatter(x, np.exp(x)); # y = e^x
```

```python
# OO API
#Prepare new data
rng = np.random.RandomState(0)
x= rng.rand(100)
y = rng.rand(100)
size = 1000*rng.rand(100)
colors = rng.rand(100)

fig, ax = plt.subplots();
img = ax.scatter(x, y, s = size, c= colors, cmap = "viridis", alpha=0.5);
fig.colorbar(img);
```

### 5. Supplots

```python
# Option 1
fix, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
# plot data each axis
ax1.plot(x, x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.random(10), np.random.random(10));

ax3.barh(y = softDrinkPrice.keys(), width= softDrinkPrice.values(), color = "green");

ax4.hist(studentHeight, bins = 30);
```

## <b> Project: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("california_cities.csv")
df.head()
```

```python
# Extrat latd = Vĩ Độ and Longd = Kinh Độ
lat, lon = df["latd"], df["longd"]
population, area = df["population_total"], df["area_total_km2"]

# plot using pyplot API
plt.figure(figsize=(10, 6))
plt.scatter(lon, lat, 
            c=np.log10(population), 
            cmap="viridis", linewidths=0,
            s = area, alpha=0.5
            );
plt.axis("equal");
plt.xlabel("longtitude");
plt.ylabel("lattitude");
plt.colorbar(label = "log$_{10}$population")

# Create legend for cities's size
areaRange = [50, 100, 300, 500]
for area in areaRange:
    plt.scatter([], [], 
                s = area, alpha=0.4, c = "k",
                label = str(area) + " km$^2$"
                );
plt.legend(labelspacing = 1, title = "City Area");
plt.title("California Cities: Population and Area Distribution");
```
